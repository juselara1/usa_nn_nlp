```python
from gpt4all import GPT4All
```

```python
model = GPT4All(
    model_name="gpt4all-falcon-q4_0.gguf",
    )
```

```python
model.generate("Alan Turing was ", max_tokens=30)
```

```python
active = True
while active:
    prompt = input()
    response = model.generate(prompt, top_k=1, max_tokens=30)
    print(response)
    if prompt == "<exit>":
        active = False
```
