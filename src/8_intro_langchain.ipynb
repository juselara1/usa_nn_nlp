```python
system_prompt = SystemMessagePromptTemplate.from_template(
    "Te voy a dar informacion en lenguaje natural, debes convertirla a el formato JSON "
    + "Te dare un texto con informacion de una persona y debes retornar UNICAMENTE "
    + "la salida en formato JSON sin ninguna expliacion adicional."
    )
```

```python
human_prompt = HumanMessagePromptTemplate.from_template(
    "Esta es la informacion: '{text}'"
)
```

```python
prompt = ChatPromptTemplate.from_messages([
    system_prompt,
    human_prompt
])
```

```python
prompt
```

```python
class JsonParser(BaseOutputParser):
    def parse(self, text: str) -> dict:
        return json.loads(text)
```

```python
chain = LLMChain(
    llm=llm,
    prompt=prompt,
    output_parser=JsonParser()
)
```

```python
response = chain.run({"text": "Maria tiene 32 anios de edad, estudio economia"})
```

```python
response.keys()
```

```python
response["edad"]
```
